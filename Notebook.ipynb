```python
## Convert DATE_TIME object to Datetime format
def datetime_format(x):
    y = pd.to_datetime(
        x,
        format = "mixed",
        dayfirst=True
    )
    return y
```

```python
plant1_gen['DATE_TIME'] = plant1_gen['DATE_TIME'].apply(datetime_format)
```

```python
plant2_gen['DATE_TIME'] = plant2_gen['DATE_TIME'].apply(datetime_format)
```

```python
plant1_sensor['DATE_TIME'] = plant1_sensor['DATE_TIME'].apply(datetime_format)
```

```python
plant2_sensor['DATE_TIME'] = plant2_sensor['DATE_TIME'].apply(datetime_format)
```

```python
# Merging Datasets with respect to their weather data
# plant 1 generation data with plant 1 waether data
merged_plant_1 = plant1_gen.merge(
    plant1_sensor,
    on='DATE_TIME',
    how='left',
    validate='many_to_one'
)

# plant 2 generation data with plant 2 sensor data
merged_plant_2 = plant2_gen.merge(
    plant2_sensor,
    on='DATE_TIME',
    how='left',
    validate='many_to_one'
)
```

```python
merged_plant_1.head()
```

```python
# Converting IRRADIATION uW/m2 to W/m2
merged_plant_1['IRRADIATION_WATT/M2'] = merged_plant_1['IRRADIATION']*1000
merged_plant_2['IRRADIATION_WATT/M2'] = merged_plant_2['IRRADIATION']*1000
```

```python
# Categorise irradiation (High,Medium,Low,No IR)
def irradiance_cat(x):
    if (x > 20) & (x <= 100):
        return("Low")
    elif (x > 100) & (x <= 700):
        return("Medium")
    elif (x > 700):
        return("High")
    else:
        return("No IR")
```

```python
merged_plant_1['IR_CAT'] = merged_plant_1['IRRADIATION_WATT/M2'].apply(irradiance_cat)
merged_plant_2['IR_CAT'] = merged_plant_2['IRRADIATION_WATT/M2'].apply(irradiance_cat)
```

```python
# Deleting irrelevant columns
merged_plant_1 = merged_plant_1.drop(columns=['PLANT_ID_x','PLANT_ID_y','SOURCE_KEY_y'],axis=1)
merged_plant_2 = merged_plant_2.drop(columns=['PLANT_ID_x','PLANT_ID_y','SOURCE_KEY_y'],axis=1)
```

```python
# Renaming Source_Key_x to panel_key
merged_plant_1 = merged_plant_1.rename(columns={'SOURCE_KEY_x':'PANEL_KEY'})
merged_plant_2 = merged_plant_2.rename(columns={'SOURCE_KEY_x':'PANEL_KEY'})
```

```python
# Adding Plant_Number Column
merged_plant_1['PLANT_NUMBER'] = "PLANT 1"
merged_plant_2['PLANT_NUMBER'] = "PLANT 2"
```

```python
# extracting time of the day
def time_of_day(x):
    hour = x.hour  
    
    if 0 <= hour < 4: return 'Late Night'
    elif 4 <= hour < 8: return 'Early Morning'
    elif 8 <= hour < 12: return 'Morning'
    elif 12 <= hour < 16: return 'Afternoon'
    elif 16 <= hour < 20: return 'Evening'
    else: return 'Night'
```

```python
merged_plant_1['TIME_OF_DAY'] = merged_plant_1['DATE_TIME'].apply(time_of_day)
merged_plant_2['TIME_OF_DAY'] = merged_plant_2['DATE_TIME'].apply(time_of_day)
```

First plant assumed panel capacity is **15000 Watts** Second plant assumed panel capacity is **1500 Watts**

```python
# Expected DC column for Plant 1
CAPACITY_1 = 15000
merged_plant_1['EXP_DC'] = CAPACITY_1*(merged_plant_1['IRRADIATION_WATT/M2']/1000)*(1+(-0.004)*(merged_plant_1['MODULE_TEMPERATURE']-25))
```

```python
# Expected DC column for Plant 2
CAPACITY_2 = 1500
merged_plant_2['EXP_DC'] = CAPACITY_2*(merged_plant_2['IRRADIATION_WATT/M2']/1000)*(1+(-0.004)*(merged_plant_2['MODULE_TEMPERATURE']-25))
```

```python
merged_plant_1['PANEL_EFFICIENCY'] = np.where(
    merged_plant_1['EXP_DC'] > 0,
    merged_plant_1['DC_POWER']/merged_plant_1['EXP_DC'],
    np.nan
)
```

```python
merged_plant_2['PANEL_EFFICIENCY'] = np.where(
    merged_plant_2['EXP_DC'] > 0,
    merged_plant_2['DC_POWER']/merged_plant_2['EXP_DC'],
    np.nan
)
```

```python
def panel_status(x):
    if np.isnan(x) or np.isinf(x):
        return "Not Measured"
    
    if x > 1.05:
        return "Over-Performing"
    elif x >= 0.95:  # Covers everything from 0.95 to 1.05
        return "Optimal"
    elif x >= 0.74:  # Covers everything from 0.74 to 0.949
        return "Under-Performing"
    elif x >= 0.50:  # Covers everything from 0.50 to 0.739
        return "Failing"
    else:            # Covers everything below 0.50
        return "Critical"
```

```python
merged_plant_1['PANEL_STATUS'] = merged_plant_1['PANEL_EFFICIENCY'].apply(panel_status)
merged_plant_2['PANEL_STATUS'] = merged_plant_2['PANEL_EFFICIENCY'].apply(panel_status)
```

```python
# Concatenating Datasets
merged = pd.concat([merged_plant_1,merged_plant_2],axis=0)
```

```python
# CSV file output
merged.to_csv("merged.csv")
```
